```python
import numpy as np
import pandas as pd
import math
```

```python
df_2021 = pd.read_csv('KCPD_Crime_Data_2021.csv')
df_2022 = pd.read_csv('KCPD_Crime_Data_2022.csv')
df_2023 = pd.read_csv('KCPD_Crime_Data_2023.csv')
```

```python
df_2021.head()
```

```python
df_2021['Offense'].unique()
```

```python
data2021 = df_2021['Offense'].value_counts().reset_index()
data2021.columns = ['Type', '2021']
data2022 = df_2022['Offense'].value_counts().reset_index()
data2022.columns = ['Type', '2022']
data2023 = df_2023['Offense'].value_counts().reset_index()
data2023.columns = ['Type', '2023']
```

```python
merged_df = pd.merge(data2021, data2022, on='Type', how='outer')
merged_df = pd.merge(merged_df, data2023, on='Type', how='outer')
```

```python
merged_df.to_csv('Kansas_212223.csv', index = False)
```

```python
df_final = pd.read_csv('kansas_final.csv')
df_final.head()
```

```python
df_final['2021'] = df_final['2021']//5
df_final['2022'] = df_final['2022']//5
df_final['2023'] = df_final['2023']//5
```

```python
df_final[['2021', '2022', '2023']] = df_final[['2021', '2022', '2023']].applymap(math.floor)
```

```python
df_final.to_csv('Kansas_final_100000.csv', index = False)
```
