```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df_Kansas = pd.read_csv('cities_crime_data/Kansas.csv')
df_Lawrence = pd.read_csv('cities_crime_data/Lawrence.csv')
df_Hutchinson = pd.read_csv('cities_crime_data/Hutchinson.csv')
df_Lenexa = pd.read_csv('cities_crime_data/Lenexa.csv')
df_Overland_park = pd.read_csv('cities_crime_data/Overland_park.csv')
df_Olathe = pd.read_csv('cities_crime_data/Olathe.csv')
df_Salina = pd.read_csv('cities_crime_data/Salina.csv')
df_Shawnee = pd.read_csv('cities_crime_data/Shawnee.csv')
df_Topeka = pd.read_csv('cities_crime_data/Topeka.csv')
df_Wichita = pd.read_csv('cities_crime_data/Wichita.csv')
```

```python
#Remove trailing zeros
from decimal import Decimal

def remove_trailing_zeros(df):
    for column in df.columns:
        if df[column].dtype == 'float64':
            df[column] = df[column].apply(lambda x: Decimal(x).normalize() if x == float(x) else x)
    return df
        
df_Kansas = remove_trailing_zeros(df_Kansas)
df_Lawrence = remove_trailing_zeros(df_Lawrence)
df_Hutchinson = remove_trailing_zeros(df_Hutchinson)
df_Lenexa = remove_trailing_zeros(df_Lenexa)
df_Overland_park = remove_trailing_zeros(df_Overland_park)
df_Olathe = remove_trailing_zeros(df_Olathe)
df_Salina = remove_trailing_zeros(df_Salina)
df_Shawnee = remove_trailing_zeros(df_Shawnee)
df_Topeka = remove_trailing_zeros(df_Topeka)
df_Wichita = remove_trailing_zeros(df_Wichita)
```

```python

# Set index for each DataFrame
df_Kansas.set_index('Type', inplace=True)
df_Lawrence.set_index('Type', inplace=True)
df_Hutchinson.set_index('Type', inplace=True)
df_Lenexa.set_index('Type', inplace=True)
df_Overland_park.set_index('Type', inplace=True)
df_Olathe.set_index('Type', inplace=True)
df_Salina.set_index('Type', inplace=True)
df_Shawnee.set_index('Type', inplace=True)
df_Topeka.set_index('Type', inplace=True)
df_Wichita.set_index('Type', inplace=True)
```

```python
# Create a dictionary of cities and their corresponding DataFrames
Cities = {
    "Kansas": df_Kansas,
    "Lawrence": df_Lawrence, 
    "Hutchinson": df_Hutchinson, 
    "Lenexa": df_Lenexa, 
    "Overland park": df_Overland_park, 
    "Olathe": df_Olathe, 
    "Salina": df_Salina, 
    "Shawnee": df_Shawnee, 
    "Topeka": df_Topeka, 
    "Wichita": df_Wichita
}
```

```python
# Define years and crime types
Years = [str(i) for i in range(2007, 2021)]
Crime_types = ["Murders", "Rapes", "Robberies", "Assaults", "Burglaries", "Thefts", "Auto thefts", "Arson"]
```

```python
# Plot data for each crime type
for Crime_type in Crime_types:
    plt.figure(figsize=(10, 6))
    for city_name, city_df in Cities.items():
        plt.plot(city_df.columns.astype(int), city_df.loc[Crime_type], marker='o', label=city_name)
        
    plt.title(f"{Crime_type} Trend in Different Cities (2007-2020)")
    plt.xlabel("Year")
    plt.ylabel(f"Incidents (per 100,000) - {Crime_type}")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
#Correlation Analysis between crimes in Kansas

#Calculate correlation matrix
correlation_matirix = df_Kansas.corr()

#Plot correlation heatmap
plt.figure(figsize = (10, 8))
sns.heatmap(correlation_matirix, annot = True, cmap = 'coolwarm', fmt = '.2f')

#Set tick labels for both axes
plt.xticks(ticks = np.arange(len(correlation_matirix.index)), labels = correlation_matirix.index, rotation=45)
plt.yticks(ticks = np.arange(len(correlation_matirix.index)), labels = correlation_matirix.index, rotation=0)

plt.title("Correlation Heatmap of Crime Types in Kansas")
plt.show()
```
